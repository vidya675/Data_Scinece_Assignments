```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
df = pd.read_csv('bank-full.csv',delimiter=';')
import warnings
warnings.filterwarnings('ignore')
import matplotlib.pyplot as plt
df.head()
```

```python
df.columns
```

```python
#Converting the character data in numerical data using LABELENCODER

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
df['job']=le.fit_transform(df['job'])
df['marital']=le.fit_transform(df['job'])
df['education']=le.fit_transform(df['education'])
df['default']=le.fit_transform(df['default'])
df['housing']=le.fit_transform(df['housing'])
df['loan']=le.fit_transform(df['loan'])
df['contact']=le.fit_transform(df['contact'])
df['month']=le.fit_transform(df['month'])
df['poutcome']=le.fit_transform(df['poutcome'])
df['y']=le.fit_transform(df['y'])
```

```python
x = df.iloc[:,:16]
y=df.iloc[:,16]
```

```python
classfier = LogisticRegression()
classfier.fit(x,y)
```

```python
y_pred = classfier.predict(x)
```

```python
y_pred
```

```python
y_pred_df = pd.DataFrame({'actual':y,'predicted_prob':classfier.predict(x)})
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix=confusion_matrix(y,y_pred)
print(confusion_matrix)
```

```python
from sklearn.metrics import  roc_curve
from sklearn.metrics import  roc_auc_score

fpr,tpr,thresholds=roc_curve(y,classfier.predict_proba(x)[:,1])

import matplotlib.pyplot as plt
plt.plot(fpr,tpr,color='red')
plt.plot([0,1],[0,1],'k--')
plt.show()
```
